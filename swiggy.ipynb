```python
import pandas as pd
```

```python
df = pd.read_csv("swiggy.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df= df.drop(columns=['id','lic_no','link','address','menu'])
```

```python
df.info()
```

```python
# First, make sure column is string type
df['city'] = df['city'].astype(str)

# Split using comma - n=1 means only first comma used (Area, City)
df[['area', 'main_city']] = df['city'].str.split(',', n=1, expand=True)

# Strip any leading/trailing white spaces
df['area'] = df['area'].str.strip()
df['main_city'] = df['main_city'].str.strip()
```

```python
df.head()
```

```python
df.drop('city', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Step 1: Replace 'Too Few Ratings' with 0
df['rating_count'] = df['rating_count'].replace('Too Few Ratings', '0')

# Step 2: Remove ' ratings' text and '+' symbol
df['rating_count'] = df['rating_count'].str.replace(' ratings', '', regex=False)
df['rating_count'] = df['rating_count'].str.replace('+', '', regex=False)

# Step 3: Convert '1K' → 1000, '2.5K' → 2500 etc.
import numpy as np

def convert_rating(value):
    value = str(value).strip()
    if 'K' in value:
        return float(value.replace('K', '')) * 1000
    else:
        return float(value)

df['rating_count'] = df['rating_count'].apply(convert_rating)
```

```python
df.head()
```

```python
# Remove ₹ and any commas
df['cost'] = df['cost'].astype(str)
df['cost'] = df['cost'].str.replace('₹', '', regex=False)
df['cost'] = df['cost'].str.replace(',', '', regex=False)

# Extract numeric part and convert to float
df['cost'] = df['cost'].str.extract(r'(\d+\.?\d*)')
df['cost'] = df['cost'].astype(float)

# Fill missing cost with median value
df['cost'] = df['cost'].fillna(df['cost'].median())
```

```python
df.head()
```

```python
# Replace '--' with 0.0 (or use 'np.nan' if you want to ignore them)
df['rating'] = df['rating'].replace('--', 0.0)

# Convert to float
df['rating'] = df['rating'].astype(float)
```

```python
df.head()
```

```python
df.info()
```

```python
# 1. Remove Duplicates
df = df.drop_duplicates()

# Drop rows with any missing values
df = df.dropna()


# 2. Handle Missing Values
df['cost'] = df['cost'].fillna(df['cost'].median())
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].median())
df['main_city'] = df['main_city'].fillna(df['area'])


# 3. Save to CSV
df.to_csv('cleaned2_swiggy_data.csv', index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics.pairwise import cosine_similarity

# Step 1: Load the full dataset and take only 3,000 rows
df = pd.read_csv('cleaned_swiggy_data.csv')
df = df.sample(n=3000, random_state=42).reset_index(drop=True)

# Step 2: Keep only necessary columns
features = df[['cuisine', 'main_city', 'area']]

# Step 3: One-hot encoding
encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(features).toarray()

# Step 4: Compute cosine similarity
similarity_matrix = cosine_similarity(encoded_features)

# Step 5: Choose a restaurant index (example: 100)
restaurant_index = 100  # You can change this value

# Step 6: Get top 5 similar restaurants
similar_indices = similarity_matrix[restaurant_index].argsort()[::-1][1:6]

# Step 7: Display recommended restaurants
print("Recommended Restaurants:")
print(df.iloc[similar_indices][['name', 'cuisine', 'area', 'main_city']])
```
